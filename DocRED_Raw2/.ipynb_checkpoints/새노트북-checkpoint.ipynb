# 각 entity별 개수 추출

```python
f = open("train.txt", "r")

ent_set = {}
rel_set = {}


for line in line_set:
    ent1, rel, ent2 = line.replace("\n", "").split("\t")
    
    try:
        ent_set[ent1] +=1
    except:
        ent_set[ent1] = 1

    try:
        ent_set[ent2] +=1
    except:
        ent_set[ent2] = 1
    try:
        rel_set[rel] +=1
    except:
        rel_set[rel] = 1
f.close()
```

# N번 등장하는 entity들이 몇개 있는지

```python
# Make ent_nums
ent_nums = {}

for k in ent_set.keys():
    try:
        ent_nums[ent_set[k]] +=1
    except:
        ent_nums[ent_set[k]] = 1

        
#View        
for k in sorted(ent_nums.keys()):
    print(f'{k} : {ent_nums[k]}')
```

```python
# for k in sorted(nums.keys()):
#     print(f'{k} : {nums[k]}')

nums = [ent_nums[k] for k in sorted(ent_nums.keys())]

from matplotlib import pyplot as plt

plt.plot(nums)
plt.xlim(0,50)
plt.show()
```

# Dataset 만들기

```python
threshold = 5
```

```python
folder = f"data/DocRED_Refine2_Over{threshold}"
```

```python
!mkdir ./../{folder}
```

```python
f = open("train.txt", "r")
length = len(f.readlines())
f = open("train.txt", "r")
fw = open(f"./../{folder}/train.txt", "w")
fv = open(f"./../{folder}/valid.txt", "w")
ft = open(f"./../{folder}/test.txt", "w")

w_ent_set = set()
w_rel_set = set()
for i, line in enumerate(f.readlines()):
    try:
        ent1, rel, ent2 = line.replace("\n", "").split("\t")

        if ent_set[ent1] >= threshold and ent_set[ent2] >= threshold:
            w_ent_set.add(ent1)
            w_ent_set.add(ent2)
            w_rel_set.add(rel)

            if i < 0.9 * length:
                fw.write(line)
            else:
                fv.write(line)
                ft.write(line)
    except:
        print(line)
        break
f.close()
fw.close()
fv.close()
ft.close()

```

```python
fw = open(f"./../{folder}/relations.dict", "w")
for i, e in enumerate(w_rel_set):
    fw.write(f'{i}\t{e}\n')
```

```python
fw = open(f"./../{folder}/entities.dict", "w")
for i, e in enumerate(w_ent_set):
    fw.write(f'{i}\t{e}\n')
fw.close()
```

```python
f_test = open("test.txt", "r")
f_train = open("train.txt", "r")


ent_test = set()
ent_train = set()

tri_test = set()
tri_train = set()

for line in f_test.readlines():
    e1,r,e2 = line.replace("\n", "").split("\t")
    ent_test.add(e1)
    ent_test.add(e2)
    tri_test.add(line)
    
    
for line in f_train.readlines():
    e1,r,e2 = line.replace("\n", "").split("\t")
    ent_train.add(e1)
    ent_train.add(e2)
    tri_train.add(line)

f_test.close()
f_train.close()
```
